```python
def is_prime(n):
    if n == 1:
        return False
    if n == 2:
        return True
    for i in range(2, n):
        if n % i == 0:
            return False
    return True
```

```python
a = np.arange(1, 26).reshape(5, 5)
a
```

```python
prima = []
for line in a:
    for item in line:
        if is_prime(item):
            prima.append(item)
            
b = np.array(prima)
b
```

```python
c = np.array([item for item in a.flatten() if is_prime(item)])
c
```
